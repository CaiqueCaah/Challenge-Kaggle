```python
import pandas as pd
```

```python
base = pd.read_csv('train.csv')
```

```python
base.drop_duplicates()
```

```python
base.head()
```

```python
aux = True
cont = 0
while aux :
    cont = cont + 1
    if cont < base.columns.size:
        if pd.isnull(base[base.columns[cont]]).sum() > 0:
            base.drop(base.columns[cont], 1, inplace = True)
            cont = cont - 1
    else:
        aux = False
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
```

```python
for i, column in enumerate (base.columns):
    if base[column].dtype == 'object':
        labelEncoder.fit(base[column])
        base[column] = labelEncoder.transform(base[column])
        
```

```python
y = base.iloc[:, 61:62].values
x = base.iloc[:, 1:61].values  
```

```python
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
scaler_y = StandardScaler()
```

```python
x = scaler_x.fit_transform(x)
y = scaler_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x, y, test_size=0.15, random_state=0)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
x_treinamento = pca.fit_transform(x_treinamento)
x_teste = pca.transform(x_teste)
componentes = pca.explained_variance_ratio_
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf', C = 10.0)
regressor.fit(x_treinamento, y_treinamento)
```

```python
score_treinamento = regressor.score(x_treinamento, y_treinamento)
score_teste = regressor.score(x_teste, y_teste)
```

```python
print(f'Score treinamento = {score_treinamento}')
print(f'Score teste = {score_teste}')
```

```python
previsoes = regressor.predict(x_teste)
```

```python
y_teste = scaler_y.inverse_transform(y_teste)
previsoes = scaler_y.inverse_transform(previsoes)
```

```python
print(f'Valor previsto {previsoes[0]}')
print(f'Valor real {y_teste[0]}')
```
